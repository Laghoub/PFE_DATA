# Probabilité selon le plat 

```python
def probPlat(plat):
    if (plat=="Lentilles"):
        return 0.1
    if (plat=="Loubia"):
        return 0.12
    if (plat=="Riz"):
        return 0.4*0.2
    if (plat=="Spaghetti"):
        return 0.3*0.2
    if (plat=="Couscous"):
        return 0.9*0.2
    else: return 0
    
```

```python
probPlat("Couscous")
```

# Probabilité selon la météo 

```python
def probMeteo(meteo):
    if (meteo=="Ensoleillé"):
        return 0.02
    if (meteo=="Nuageux"):
        return 0.06
    if (meteo=="Mauvais"):
        return 0.09
    
```

# Probabilité selon la bourse

```python
def probBourse(bourse):
    if(bourse=="Oui"):
        return 0.7*0.1
    if(bourse=="Non"):
        return 0.6*0.1
```

# Probabilité selon l'hébergement 

```python
def probHeberg(heberg):
    if(heberg=="externe"):
        return 0.25*0.8
    if(heberg=="interne"):
        return 0.25*0.5
```

# probabilité selon les revenues des parents

```python
def probRevenues(revenu):
    if(revenu<50000):
        return 0.9*0.35
    if(revenu>50000 and revenu<100000):
        return 0.6*0.35
    if (revenu>100000 and revenu<150000):
        return 0.3*0.35
    if (revenu>150000):
        return 0.05*0.35
```

# probabilité selon les evenements 

```python
def probEvent(event):
    if(event=="Normal"):
        return 1
    return 0
```

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    acf1 = acf(fc-test)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'acf1':acf1, 
            'corr':corr, 'minmax':minmax})

forecast_accuracy(fc, test.values)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator

n_input = 12
n_features= 1
generator = TimeseriesGenerator(scaled_train_data, scaled_train_data, length=n_input, batch_size=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

lstm_model = Sequential()
lstm_model.add(LSTM(200, activation='relu', input_shape=(n_input, n_features)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')

lstm_model.summary()
```

```python
lstm_model.fit_generator(generator,epochs=12)
```
